# Part 1 - Lotto.txt

```python
import numpy as np
```

```python
#1. Read the file using loadtxt()
arr1 = np.loadtxt('Lotto.txt',dtype=int)
arr1
```

```python
#2. Sort each row, from smallest number to largest
arr1.sort()
arr1
```

```python
#3. Save the sorted array to a file called lotto_sorted.txt
np.savetxt('lotto_sorted.txt',arr1,fmt='%0f')
```

```python
#4. Print to the screen all unique numbers in the array and the count of each unique number
np.unique(arr1,return_counts = True)
```

```python
#Make the results more readable through a dictionary format
k,v = np.unique(arr1,return_counts = True)
dict1 = dict(zip(k,v))
dict1
```
